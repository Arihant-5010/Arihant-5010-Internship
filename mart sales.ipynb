```python
# Check for duplicates
print(data.duplicated().sum())
data = data.drop_duplicates()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Selecting numerical columns
numerical_cols = data.select_dtypes(include=['float64', 'int64']).columns

# Boxplot for all numerical columns
plt.figure(figsize=(12, 8))
sns.boxplot(data=data[numerical_cols], palette='pastel')

# Add title and labels
plt.title('Boxplots of Numerical Columns', fontsize=16)
plt.ylabel('Values', fontsize=12)
plt.xlabel('Columns', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
import pandas as pd

# Define a function to remove outliers using IQR
def remove_outliers(data, column):
    Q1 = data[column].quantile(0.25)  # First quartile (25th percentile)
    Q3 = data[column].quantile(0.75)  # Third quartile (75th percentile)
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR  # Lower whisker
    upper_bound = Q3 + 1.5 * IQR  # Upper whisker
    
    # Filter out rows that are outliers
    data = data[(data[column] >= lower_bound) & (data[column] <= upper_bound)]
    return data

# Remove outliers from 'Sales' and 'Profit'
data = remove_outliers(data, 'Sales')
data = remove_outliers(data, 'Profit')

# Check if outliers are removed
print(data.describe())
```

```python
# Ensure Order Date is in datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Group by month and calculate total sales
monthly_sales = data.resample('M', on='Order Date')['Sales'].sum()

plt.figure(figsize=(10, 5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='blue')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(linestyle='--', alpha=0.7)
plt.show()
```

```python
# Group data by Region and Segment
region_segment_sales = data.groupby(['Region', 'Segment'])['Sales'].sum().unstack()

region_segment_sales.plot(kind='bar', stacked=True, figsize=(10, 6), color=['#FFA07A', '#20B2AA', '#778899'])
plt.title('Sales by Region and Segment')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.legend(title='Segment')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.show()
```

```python
# Get top 10 products by sales
top_products = data.groupby('Product Name')['Sales'].sum().nlargest(10)

top_products.plot(kind='barh', figsize=(8, 5), color='teal', alpha=0.8)
plt.title('Top 10 Products by Sales')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Ensure 'Order Date' is in datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Resample data by month to calculate total Sales and Profit
monthly_data = data.resample('M', on='Order Date')[['Sales', 'Profit']].sum()

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(monthly_data.index, monthly_data['Sales'], label='Sales', marker='o', linestyle='-', color='blue')
plt.plot(monthly_data.index, monthly_data['Profit'], label='Profit', marker='o', linestyle='--', color='orange')

# Add titles and labels
plt.title('Monthly Sales and Profit Trends', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.axhline(0, color='gray', linestyle='--', linewidth=1)  # Optional: Add a baseline
plt.legend(title='Metric', fontsize=10)
plt.grid(linestyle='--', alpha=0.5)

# Display the plot
plt.show()
```

```python
# Extract month from 'Order Date' for seasonality analysis
data['Month'] = data['Order Date'].dt.month

# Group by Month and calculate average Sales and Profit
seasonal_data = data.groupby('Month')[['Sales', 'Profit']].mean()

# Plot seasonal trends
plt.figure(figsize=(10, 6))
plt.plot(seasonal_data.index, seasonal_data['Sales'], label='Average Sales', marker='o', linestyle='-', color='green')
plt.plot(seasonal_data.index, seasonal_data['Profit'], label='Average Profit', marker='o', linestyle='--', color='purple')

# Add titles and labels
plt.title('Seasonal Trends in Sales and Profit', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Average Value', fontsize=12)
plt.xticks(ticks=range(1, 13), labels=[
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Metric', fontsize=10)
plt.grid(linestyle='--', alpha=0.5)

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group data by 'Region' and calculate total Sales and Profit
region_data = data.groupby('Region')[['Sales', 'Profit']].sum().reset_index()

# Create a grouped bar chart
fig, ax = plt.subplots(figsize=(10, 6))

bar_width = 0.35
x = range(len(region_data))

# Plot Sales and Profit side by side
ax.bar(x, region_data['Sales'], width=bar_width, label='Sales', color='skyblue')
ax.bar([i + bar_width for i in x], region_data['Profit'], width=bar_width, label='Profit', color='orange')

# Add labels and titles
ax.set_xlabel('Region', fontsize=12)
ax.set_ylabel('Amount', fontsize=12)
ax.set_title('Sales and Profit by Region', fontsize=14)
ax.set_xticks([i + bar_width / 2 for i in x])
ax.set_xticklabels(region_data['Region'])
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
